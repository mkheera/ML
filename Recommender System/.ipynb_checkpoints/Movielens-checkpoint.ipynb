MovieLens 1M Dataset
SUMMARY
================================================================================

These files contain 1,000,209 anonymous ratings of approximately 3,900 movies 
made by 6,040 MovieLens users who joined MovieLens in 2000.

RATINGS FILE DESCRIPTION
================================================================================

All ratings are contained in the file "ratings.dat" and are in the
following format:

UserID::MovieID::Rating::Timestamp

- UserIDs range between 1 and 6040 
- MovieIDs range between 1 and 3952
- Ratings are made on a 5-star scale (whole-star ratings only)
- Timestamp is represented in seconds since the epoch as returned by time(2)
- Each user has at least 20 ratings

USERS FILE DESCRIPTION
================================================================================

User information is in the file "users.dat" and is in the following
format:

UserID::Gender::Age::Occupation::Zip-code

All demographic information is provided voluntarily by the users and is
not checked for accuracy.  Only users who have provided some demographic
information are included in this data set.

- Gender is denoted by a "M" for male and "F" for female
- Age is chosen from the following ranges:

	*  1:  "Under 18"
	* 18:  "18-24"
	* 25:  "25-34"
	* 35:  "35-44"
	* 45:  "45-49"
	* 50:  "50-55"
	* 56:  "56+"

- Occupation is chosen from the following choices:

	*  0:  "other" or not specified
	*  1:  "academic/educator"
	*  2:  "artist"
	*  3:  "clerical/admin"
	*  4:  "college/grad student"
	*  5:  "customer service"
	*  6:  "doctor/health care"
	*  7:  "executive/managerial"
	*  8:  "farmer"
	*  9:  "homemaker"
	* 10:  "K-12 student"
	* 11:  "lawyer"
	* 12:  "programmer"
	* 13:  "retired"
	* 14:  "sales/marketing"
	* 15:  "scientist"
	* 16:  "self-employed"
	* 17:  "technician/engineer"
	* 18:  "tradesman/craftsman"
	* 19:  "unemployed"
	* 20:  "writer"

MOVIES FILE DESCRIPTION
================================================================================

Movie information is in the file "movies.dat" and is in the following
format:

MovieID::Title::Genres

- Titles are identical to titles provided by the IMDB (including
year of release)
- Genres are pipe-separated and are selected from the following genres:

	* Action
	* Adventure
	* Animation
	* Children's
	* Comedy
	* Crime
	* Documentary
	* Drama
	* Fantasy
	* Film-Noir
	* Horror
	* Musical
	* Mystery
	* Romance
	* Sci-Fi
	* Thriller
	* War
	* Western

- Some MovieIDs do not correspond to a movie due to accidental duplicate
entries and/or test entries
- Movies are mostly entered by hand, so errors and inconsistencies may exist

# 1. A simple Recommender System based on popularity

```python
# grouping movies by id to find sum of ratings and no of votes for each movie
r=ratings.groupby(['movie_id',]).agg({'user_id': 'count','rating':'sum'}).reset_index().rename(columns={'user_id':'votecount','rating':'votesum'})
#c=r.sort_values('rating', ascending=False).head(20)
```

######  I am using IMDB rating formula(source wikipedia) to implement Popularity based RC.

The formula for calculating the Top Rated 250 Titles:
weighted rating (WR) = (v ÷ (v+m)) × R + (m ÷ (v+m)) × C
Where:
R = average for the movie (mean) = (Rating)
v = number of votes for the movie = (votes)
m = minimum votes required to be listed in the Top 250 (currently 25000)
C = the mean vote across the whole report (currently 7.0)

```python
#average movie rating is 3.581 on a scale of 5
#i am using votecount(m) as 10000  to get a listing of around 200 movies
C=ratings['rating'].mean()

#filter movies with count of rating more than 1000
m=1000
s_movies=r[r['votecount']>=m].copy()
s_movies.shape
```

```python

s_movies['vote_average']=s_movies['votesum']/s_movies['votecount']
#poprs=s_movies.sort_values('vote_average',ascending=False)
s_movies.head(5)
```

```python
def weighted_rating(x,m=m,C=C):
    v=x['votecount']
    R=x['vote_average']
    return (v/(v+m)*R)+(m/(m+v)*C)
```

```python
#calculate score for all movies and sort them
s_movies['score']=s_movies.apply(weighted_rating,axis=1)
```

```python
s_movies.sort_values('score',ascending=False,inplace=True)
s_movies.head(5)
```

# 2. Content Based Recommender System using Genre.

```python
#trying to find the distribution of movies on the basis of genre.
from collections import Counter
Counter(a).keys() # equals to list(set(words))
Counter(a).items() # counts the elements' frequency
```

```python
df=pd.Series(movies['genre'].str.split('|')).apply(frozenset).to_frame(name='genre')
```

```python
#Encoding genre 
for genre in frozenset.union(*df.genre):
    df[genre]=df.apply(lambda _:int(genre in _.genre),axis=1)
```

```python
df
```

```python
df.drop(['genre'],axis=1,inplace=True)
```

```python
#computing similarity matrix(of genre) for each movie wih every other movie using cosine_similarity
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim=cosine_similarity(df)

#There are 3883 movies and hence we get a similarity matrix of size 3883*3883
```

```python
#creating a series with title as index to make it easy to get ids of movies based on title
titles=movies['title']
ind=pd.Series(movies.index,index=movies['title'])
```

```python
#function to find and return movies based on genre similarity
def conRS(title):
    id=ind[title]
    sim_score=list(enumerate(cosine_sim[id]))
    sim_score=sorted(sim_score,key=lambda x:x[1],reverse=True)
    sim_score=sim_score[1:100]
    mov_indx=[i[0] for i in sim_score]
    return titles.iloc[mov_indx]
```

# 3. Collaborative Filtering RS

Collaborative Filtering RS
Implementation.
we need only the ratings(user_id,movie_id,rating) file to make recommendations.
1. user-user collaborative filtering-we find similar users by similarity in ratings given for movies
2.item_item collaborative filtering-we find similar movies by looking at the similarity in ratings 

```python
#Here i am choosing 100k data because of limitation in my laptop's computing power
data=ratings[:100000]
dat=data[0:4]
data.shape
```

```python
n_users = data.user_id.unique().shape[0]
n_movies = data.movie_id.unique().max()
print('Number of users = ' + str(n_users) + ' | Number of movies = ' + str(n_movies))

```

```python
#Sparsity of the 100k data sample
sparsity=round(1.0-len(ratings)/float(n_users*n_movies),3)
print('The sparsity level of Movielens data sample is ' +  str(sparsity*100) + '%')
```

```python
#from sklearn import cross_validation as cv
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, test_size=0.3)
```

```python
#Create two user-item matrices, one for training and another for testing
train_data_matrix = np.zeros((n_users, n_movies))
for line in train_data.itertuples():
    train_data_matrix[line[1]-1, line[2]-1] = line[3]

test_data_matrix = np.zeros((n_users, n_movies))
for line in test_data.itertuples():
    test_data_matrix[line[1]-1, line[2]-1] = line[3]
```

```python
from sklearn.metrics.pairwise import pairwise_distances
user_similarity = 1-pairwise_distances(train_data_matrix, metric='cosine')
item_similarity = 1-pairwise_distances(train_data_matrix.T, metric='cosine')
```

```python
def predict(ratings, similarity, type='user'):
    if type == 'user':
        mean_user_rating = ratings.mean(axis=1)
        #You use np.newaxis so that mean_user_rating has same format as ratings
        ratings_diff = (ratings - mean_user_rating[:, np.newaxis])
        pred = mean_user_rating[:, np.newaxis] + similarity.dot(ratings_diff) / np.array([np.abs(similarity).sum(axis=1)]).T
    elif type == 'item':
        pred = ratings.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])
    return pred
```

```python

item_prediction = predict(train_data_matrix, item_similarity, type='item')
user_prediction = predict(train_data_matrix, user_similarity, type='user')
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
def rmse(prediction, trueval):
    prediction = prediction[trueval.nonzero()].flatten()
    trueval = trueval[trueval.nonzero()].flatten()
    return sqrt(mean_squared_error(prediction, trueval))
```

```python
print('User-based CF RMSE: ' + str(rmse(user_prediction, test_data_matrix)))
print('Item-based CF RMSE: ' + str(rmse(item_prediction, test_data_matrix)))
```

```python
def recommend_movies(predictions_df, userID, movies_df, original_ratings_df, num_recommendations=5):
    
    # Get and sort the user's predictions
    user_row_number = userID - 1 # UserID starts at 1, not 0
    sorted_user_predictions = predictions_df.iloc[user_row_number].sort_values(ascending=False)
    
    # Get the user's data and merge in the movie information.
    user_data = ratings[ratings.user_id == (userID)]
    user_full = (user_data.merge(movies, how = 'left', left_on = 'movie_id', right_on = 'movie_id').
                     sort_values(['rating'], ascending=False)
                 )

    print ('User {0} has already rated {1} movies.'.format(userID, user_full.shape[0]))
    print ('Recommending the highest {0} predicted ratings movies not already rated.'.format(num_recommendations))
    
    # Recommend the highest predicted rating movies that the user hasn't seen yet.
    recommendations = (movies_df[~movies_df['movie_id'].isin(user_full['movie_id'])].
         merge(pd.DataFrame(sorted_user_predictions).reset_index(), how = 'left',
               left_on = 'movie_id',
               right_on = 'movie_id').
         rename(columns = {user_row_number: 'Predictions'}).
         sort_values('Predictions', ascending = False).
                       iloc[:num_recommendations, :-1]
                      )

    return user_full, recommendations
```

###### Item-Item Collaborative Filtering using KNN

```python
from scipy.sparse import csr_matrix
#from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.neighbors import NearestNeighbors
#rating matrix as pivot table of user_id/movie_id
r_mat = ratings.pivot(index = 'user_id', columns ='movie_id', values = 'rating').fillna(0)
r_mat_sparse = csr_matrix(r_mat.values)  #for user-user CF using KNN
```

```python
#pivot table for Item-item CF 
m_mat=r_mat.T
m_mat_sparse=csr_matrix(m_mat.values)
```

```python
#algorithm : {‘auto’, ‘ball_tree’, ‘kd_tree’, ‘brute’}, optional
#using euclidean distance to calculate similarity
model_knn = NearestNeighbors(metric = 'euclidean', algorithm = 'auto')
model_knn.fit(m_mat_sparse)
```

```python
#Finding movies similar to Movie 1(Toy story)
query_item=1   #( 1)
query_index=query_item-1
distances, indices = model_knn.kneighbors(m_mat.iloc[query_index, :].values.reshape(1, -1), n_neighbors = 10)

for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for user {0}:\n'.format(m_mat.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, m_mat.index[indices.flatten()[i]], distances.flatten()[i]))
```
